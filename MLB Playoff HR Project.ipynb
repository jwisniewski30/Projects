```python
# Stadium Analysis

import matplotlib.pyplot as plt

# Data
venues = ["Target Field", "Rogers Centre", "Minute Maid Park"]
years = [2018, 2019, 2020, 2021, 2022]
hr_data = {
    "Target Field": [92, 95, 86, 110, 97],
    "Rogers Centre": [100, 112, 123, 101, 117],
    "Minute Maid Park": [99, 124, 81, 99, 112],
}

# Create a bar chart for each venue
plt.figure(figsize=(10, 6))

for venue in venues:
    plt.plot(years, hr_data[venue], marker='o', label=venue)

plt.title('Home Run Frequency by Venue (2018-2022)')
plt.xlabel('Year')
plt.ylabel('Home Runs')
plt.legend()
plt.grid(True)
plt.xticks(years)
plt.show()
```

```python
# Data
venues = ["Target Field", "Rogers Centre", "Minute Maid Park"]
averages = [96, 110.6, 103]  # Average frequency of home runs for each venue

# Create a bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(venues, averages, color='green')
plt.title('Average Frequency of Home Runs by Venue')
plt.xlabel('Venue')
plt.ylabel('Average')
plt.ylim(0, max(averages) + 10)  # Adjust the y-axis limit for better visualization
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add labels with actual values to each bar
for bar, avg in zip(bars, averages):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.15, bar.get_height() + 2, str(avg), fontsize=12)

plt.show()
```
