```python
import pandas as pd

# Define path to Excel file.
file_path = 'Consolidated Sites.xlsx'

# Use the read_excel function to import the Excel file into a DataFrame
df = pd.read_excel(file_path)
```

```python
import matplotlib.pyplot as plt

# Filter only the relevant columns (Site and Current 3 Month Avg).
subset_df = df[['Site', 'Current 3 month Avg']]

# Sort the DataFrame by 'Current 3 month Avg' in ascending order.
subset_df = subset_df.sort_values(by='Current 3 month Avg')

# Create a bar chart with green bars.
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed.
plt.bar(subset_df['Site'], subset_df['Current 3 month Avg'], color='green')  # Set the bar color to green.
plt.ylabel('Score')
plt.title('Current 3 Month Averages by Site (Sorted)')
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility.

# Show the plot.
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
file_path = 'Little Rock DPI.xlsx'
df = pd.read_excel(file_path, engine='openpyxl')

# Display basic statistics
print("Basic Statistics:")
print(df.describe())

# Calculate average score by LOB
average_scores_by_lob = df.groupby('LOB')['Score'].mean()

# Create a bar plot for average scores by LOB
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.barplot(x=average_scores_by_lob.index, y=average_scores_by_lob.values)
plt.title("Average Score by LOB")
plt.xlabel("LOB")
plt.ylabel("Average Score")
plt.xticks(rotation=45)
plt.show()

# Create a count plot for medals distribution by LOB
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='LOB', hue='Medal', order=df['LOB'].value_counts().index)
plt.title("Medals Distribution by LOB")
plt.xlabel("LOB")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title='Medal')
plt.show()
```

```python
import pandas as pd

# Read the CSV file into a Pandas DataFrame
data = pd.read_csv('Drivers Report.csv')

# Display the first few rows of the DataFrame to ensure it's loaded correctly
print(data.head())

# Basic statistics for the 'Coachable Score_Total' column
coachable_score_stats = data['Coachable Score_Total'].describe()
print("Coachable Score Statistics:")
print(coachable_score_stats)

# Basic statistics for the 'Total Score_Total' column
total_score_stats = data['Total Score_Total'].describe()
print("\nTotal Score Statistics:")
print(total_score_stats)

# Finding the person with the highest 'Coachable Score_Total'
max_coachable_score = data[data['Coachable Score_Total'] == data['Coachable Score_Total'].max()]
print("\nPerson with the Highest Coachable Score:")
print(max_coachable_score)

# Finding the person with the highest 'Total Score_Total'
max_total_score = data[data['Total Score_Total'] == data['Total Score_Total'].max()]
print("\nPerson with the Highest Total Score:")
print(max_total_score)
```

```python
import matplotlib.pyplot as plt

# Visualization 1: Histogram for Coachable Score_Total
plt.figure(figsize=(8, 6))
plt.hist(data['Coachable Score_Total'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Coachable Score_Total')
plt.xlabel('Coachable Score_Total')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Visualization 2: Histogram for Total Score_Total
plt.figure(figsize=(8, 6))
plt.hist(data['Total Score_Total'], bins=20, color='lightcoral', edgecolor='black')
plt.title('Distribution of Total Score_Total')
plt.xlabel('Total Score_Total')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd

# Input path to Excel file
file_path = 'Survey.xlsx'

# Load the Excel file into a DataFrame
df = pd.read_excel(file_path)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
import matplotlib.pyplot as plt

# How long have you been employed as a WM Driver?
employee_tenure_counts = df['How long have you been employed as a WM Driver?'].value_counts()

# As a WM Driver, what LOB do you primarily support?
lob_counts = df['As a WM Driver, what LOB do you primarily support?'].value_counts()

# How often do you receive feedback from your Route Manager?
feedback_counts = df['How often do you receive feedback from your Route Manager?'].value_counts()

# What is your relationship with your Route Manager?
relationship_counts = df['What is your relationship with your Route Manager?'].value_counts()

# What additional support do you need from your Route Manager to help you be successful in your role?
support_needs = df['What additional support do you need from your Route Manager to help you be successful in your role? (Select all that apply)'].str.split(';')
support_needs = [item for sublist in support_needs for item in sublist]
support_needs_counts = pd.Series(support_needs).value_counts()

# What are your primary reasons for calling your RM while on route?
call_reasons = df['What are your primary reasons for calling your RM while on route?'].str.split(';')
call_reasons = [item for sublist in call_reasons for item in sublist]
call_reasons_counts = pd.Series(call_reasons).value_counts()

# Plotting the data
employee_tenure_counts.plot(kind='bar', title="WM Driver Tenure")
plt.xlabel("Tenure")
plt.ylabel("Count")
plt.show()

lob_counts.plot(kind='bar', title="LOB Support")
plt.xlabel("LOB")
plt.ylabel("Count")
plt.show()

feedback_counts.plot(kind='bar', title="Feedback Frequency")
plt.xlabel("Feedback Frequency")
plt.ylabel("Count")
plt.show()

relationship_counts.plot(kind='bar', title="Relationship with RM")
plt.xlabel("Relationship")
plt.ylabel("Count")
plt.show()

support_needs_counts.plot(kind='bar', title="Additional Support Needs")
plt.xlabel("Support Needs")
plt.ylabel("Count")
plt.show()

call_reasons_counts.plot(kind='bar', title="Call Reasons")
plt.xlabel("Call Reasons")
plt.ylabel("Count")
plt.show()
```
